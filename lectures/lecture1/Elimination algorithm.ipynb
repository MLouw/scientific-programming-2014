We are now working on an early classification dataset called *zoo*. 

```python
import codecs
import pandas as pd
from sklearn.feature_extraction import DictVectorizer
```

```python
attribute_names = "animal_name hair feathers eggs milk airborne aquatic predator toothed backbone breathes venomous fins legs tail domestic catsize type".split(" ")

type1 = []
labels = []
names = []
data = []
for line in codecs.open("zoo.data"):
    attribute_values =  line.strip().split(",")
    attributes = dict(zip(attribute_names, attribute_values))
#    print attributes['type'] + "\t",
    # save and remove type from feature vector
    labels.append(int(attributes['type']))
    del attributes['type']
    
    instance = {}
    data.append(instance)

    for key, val in attributes.items():
        # store name
        if key == 'animal_name':
            names.append(val)
        # binarize information on number of legs
        elif key == 'legs' and int(val) >= 1:
            instance[key + val] = 1

        else:
            instance[key] = int(val)
            #instance['name'] = val
            #print val,
#        elif int(val) > 0:
#            if key == 'legs':
#                print key + str(val),
#            else:
#                print key,
del attributes['legs']
                

    
vectorizer = DictVectorizer(sparse=False)
X = vectorizer.fit_transform(data)
D = pd.DataFrame(X, columns=vectorizer.get_feature_names(), index=names, dtype=int)

labels = pd.Series(labels, index=names)

print labels

labels.to_csv("animals.labels")
D.to_csv("animals.data")

#for line codecs.open("zoo.data"):
#    line.strip()
```

Now that we have our data in place, we can run our algorithm on it to find groups that share the same features.

```python
def find_common_traits(label):
    common = None
    for name, row in D[labels == label].iterrows():
        attributes = [key for (key, val) in row.to_dict().items() if val]
        if not common:
            common = set(attributes)
        else:
            common = common & set(attributes)
    return common

for label in sorted(labels.unique().tolist()):
    print "Group", label, find_common_traits(label)
    
    #print row.to_dict()
```
